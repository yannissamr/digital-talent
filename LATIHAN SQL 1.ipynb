```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SHOW TABLES")
```

```python
#1. Prepare a list of offices sorted by country, state, city.

run("SELECT country, state, city FROM offices")
```

```python
#2. How many employees are there in the company?

run("SELECT COUNT(*) AS 'totalEmployees' FROM employees ")
```

```python
#3. What is the total of payments received?

run("SELECT SUM(amount) AS 'totalPayment' FROM payments")
```

```python
#4. List the product lines that contain 'Cars

run("SELECT * FROM productlines")
```

```python
run("SELECT * FROM productlines WHERE productLine LIKE '%car%'")
```

```python
#5. Report total payments for October 28, 2004.

run("""
SELECT paymentDate, 
SUM(amount) AS totalPayments 
FROM payments 
WHERE paymentDate='2004-10-28'
""")
```

```python
#6. Report those payments greater than $100,000.

run("SELECT * FROM payments WHERE amount > 100000")
```

```python
#7. List the products in each product line.

run("""
SELECT productCode,productName,productLine 
FROM products
ORDER BY productLine
""")
```

```python
#8. How many products in each product line?

run("SELECT productLine, COUNT(*) AS totalProduct FROM products GROUP BY productLine")
```

```python
#9. What is the minimum payment received?

run("SELECT MIN(amount) FROM payments")
```

```python
#10. List all payments greater than twice the average payment.

run("SELECT AVG(amount) FROM payments")
```

```python
run("SELECT * FROM payments WHERE amount>(SELECT AVG(amount) FROM payments)*2")
```

```python
#11. What is the average percentage markup of the MSRP on buyPrice?

run("SELECT AVG((MSRP-buyPrice)/buyPrice*100) AS avgPercentageMarkup FROM products") 
```

```python
#12. How many distinct products does ClassicModels sell?

run("SELECT DISTINCT productName AS ClassicModelsName FROM products")
```

```python
run("SELECT COUNT(DISTINCT productName) AS totalSellClassicModels FROM products")
```

```python
#13. Report the name and city of customers who don't have sales representatives?

run("SELECT customerName,city FROM customers WHERE salesRepEmployeeNumber IS NULL")
```

```python
#14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

run("""
SELECT 
    CONCAT(firstName,' ',lastName) AS fullName, jobTitle
FROM employees 
WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
""")
```

```python
#15. Which orders have a value greater than $5,000?

run("""
SELECT orderNumber,sum(priceEach * quantityOrdered) AS totalValue
FROM orderDetails
GROUP BY orderNumber
HAVING totalValue>5000
""")
```
