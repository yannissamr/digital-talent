# Deskripsi Data

```python
import pandas as pd

print('Berikut ini adalah data jumlah kunjungan wisatawan pada tahun 2008-2012 berdasarkan lokasi daya tarik wisata:')
data_kunjungan_wisata = pd.read_csv('kunjunganwisatajakarta.csv', delimiter = ',', encoding = "ISO-8859-1")

data_kunjungan_wisata
```

# Statistik Data

```python
# Jumlah total kunjungann wisata dari tahun 2008-2012
print("Jumlah total kunjungann wisata dari tahun 2008-2012 adalah", data_kunjungan_wisata.jumlah_kunjungan.sum())

# Jumlah total kunjungan wisata tahun 2008
print("Jumlah total kunjungan wisata pada tahun 2008 adalah", data_kunjungan_wisata[data_kunjungan_wisata['tahun'] == 2008].jumlah_kunjungan.sum())

# Jumlah total kunjungan wisata tahun 2009
print("Jumlah total kunjungan wisata pada tahun 2009 adalah", data_kunjungan_wisata[data_kunjungan_wisata['tahun'] == 2009].jumlah_kunjungan.sum())

# Jumlah total kunjungan wisata tahun 2010
print("Jumlah total kunjungan wisata pada tahun 2010 adalah", data_kunjungan_wisata[data_kunjungan_wisata['tahun'] == 2010].jumlah_kunjungan.sum())

# Jumlah total kunjungan wisata tahun 2011
print("Jumlah total kunjungan wisata pada tahun 2011 adalah", data_kunjungan_wisata[data_kunjungan_wisata['tahun'] == 2011].jumlah_kunjungan.sum())

# Jumlah total kunjungan wisata tahun 2012
print("Jumlah total kunjungan wisata pada tahun 2012 adalah", data_kunjungan_wisata[data_kunjungan_wisata['tahun'] == 2012].jumlah_kunjungan.sum())
```

```python
dtw1 = data_kunjungan_wisata['lokasi'] == 'Taman Impian Jaya Ancol'
dtw2 = data_kunjungan_wisata['lokasi'] == 'TMII'
dtw3 = data_kunjungan_wisata['lokasi'] == 'Ragunan'
dtw4 = data_kunjungan_wisata['lokasi'] == 'Museum Satria Mandala'
dtw5 = data_kunjungan_wisata['lokasi'] == 'Museum Sejarah Jakarta'
dtw6 = data_kunjungan_wisata['lokasi'] == 'Pelabuhan Sunda Kelapa'
tahun1 = data_kunjungan_wisata['tahun'] == 2008
tahun2 = data_kunjungan_wisata['tahun'] == 2009
tahun3 = data_kunjungan_wisata['tahun'] == 2010
tahun4 = data_kunjungan_wisata['tahun'] == 2011
tahun5 = data_kunjungan_wisata['tahun'] == 2012

print('Jumlah kunjungan Museum Satria Mandala pada tahun 2011 adalah', data_kunjungan_wisata[dtw5 & tahun4].sum().jumlah_kunjungan)
```

```python
# Jumlah kunjungan terendah pada tahun tertentu dari seluruh lokasi
print('Jumlah kunjungan terendah pada tahun 2010 adalah sejumlah', data_kunjungan_wisata.jumlah_kunjungan.min())

# Jumlah kunjungan terendah pada lokasi tertentu dari seluruh tahun
print('Jumlah kunjungan terendah di Pelabuhan Sunda Kelapa adalah sejumlah', data_kunjungan_wisata[dtw6].jumlah_kunjungan.min())
```

```python
# Jumlah kunjungan tertinggi pada tahun tertentu dari seluruh lokasi
print('Jumlah kunjungan tertinggi pada tahun 2010 adalah sejumlah', data_kunjungan_wisata.jumlah_kunjungan.max())

# Jumlah kunjungan tertinggi pada tahun tertentu dari lokasi tertentu
print('Jumlah kunjungan tertinggi di Pelabuhan Sunda Kelapa adalah sejumlah', data_kunjungan_wisata[dtw6].jumlah_kunjungan.min())
```

```python
# Mean kunjungan wisata dari seluruh data
print('Mean kunjungan wisata di Jakarta adalah', data_kunjungan_wisata.jumlah_kunjungan.mean())

# Mean kunjungan wisata pertahun
print('Mean kunjungan wisata per tahun di Jakarta adalah', data_kunjungan_wisata[tahun2].jumlah_kunjungan.mean())

# Mean kunjungan wisata per lokasi
print('Mean kunjungan wisata per lokasi di Jakarta adalah', data_kunjungan_wisata[dtw4].jumlah_kunjungan.mean())
```

```python
# Median kunjungan wisata dari seluruh data
print('Mean kunjungan wisata di Jakarta adalah', data_kunjungan_wisata.jumlah_kunjungan.median())

# Median kunjungan wisata pertahun
print('Mean kunjungan wisata per tahun di Jakarta adalah', data_kunjungan_wisata[tahun2].jumlah_kunjungan.median())

# Median kunjungan wisata per lokasi
print('Mean kunjungan wisata per lokasi di Jakarta adalah', data_kunjungan_wisata[dtw4].jumlah_kunjungan.median())
```

```python
# Grup data kunjungan sesuai lokasi
data_kunjungan_wisata.groupby(['lokasi', 'tahun']).sum()
```

# Visualisasi Sederhana

```python
# Chart 1

ax = data_kunjungan_wisata.groupby('lokasi').sum()['jumlah_kunjungan']
ax = ax.plot(kind='bar', title="Jumlah Kunjungan Wisata Tahun 2008-2012 berdasarkan Lokasi", figsize=(10,10), legend=True, fontsize=11)
ax.set_xlabel("Lokasi Daya Tarik Wisata", fontsize=12)
ax.set_ylabel("Jumlah kunjungan wisata", fontsize=12)
```

```python
# Chart 2

ax = data_kunjungan_wisata.groupby('tahun').mean()['jumlah_kunjungan']
ax = ax.plot(kind='bar', title="Jumlah Kunjungan Wisata Bedasarkan Tahun", figsize=(10,10), legend=True, fontsize=11)
ax.set_xlabel("Lokasi Daya Tarik Wisata", fontsize=12)
ax.set_ylabel("Jumlah kunjungan wisata", fontsize=12)
```
